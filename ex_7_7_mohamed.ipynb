```
import math
import random
import scipy as sp
import numpy as np
from scipy.stats import expon
import matplotlib.pyplot as plt
%matplotlib inline

random.seed(42)
np.random.seed(42)
```

```
def X(theta):
  lambd = 1.0 / (50*theta)
  return random.expovariate(lambd)
```

```
plt.figure(figsize=(16, 8))
plt.hist([X(100) for i in range(10000)], bins=32)
```

```
thetas = np.linspace(0, 100, 1000)
Xs = np.zeros_like(thetas)
for i in range(1000):
  Xs += np.array([X(theta) for theta in thetas])
Xs /= 1000
plt.figure(figsize=(16, 8))
plt.plot(thetas, Xs)
```

# Derivative estimation

```
def G_IPA(theta):
  X_ = X(theta) # sample value from the distribution
  grad = (X_ / theta) * (1/math.sqrt(1+X_) - X_/(2*((1+X_)**(3/2))))
  return grad
```

```
def G_SF(theta):
  X_ = X(theta)
  grad = (1/(50*theta))*(X_/math.sqrt(X_ + 1))*(X_/(50*theta) - 1)
  return grad
```

```
def G_MVD(theta):
  X_ = X(theta)
  X__ = X(theta)
  grad = (50/theta)*((X_+X__)/math.sqrt(X_ + X__ + 1) - X_/math.sqrt(X_ + 1))
  return grad
```

# Confidence intervals

```
def estimate(estimator, N, theta):
  samples = []
  for n in range(N):
    samples.append(estimator(theta))
  mean = np.mean(samples)
  std = np.std(samples)
  return mean, std
```

```
N = 10000
for estimator in [G_IPA, G_SF, G_MVD]:
  for theta in [5, 8, 10, 12, 15]:
    time = %timeit estimate(estimator, N, theta)
    mean, std = estimate(estimator, N, theta)

    print('estimator: {} theta: {} mean:{} std:{} +/- {}'.format(estimator, theta, mean, std, 1.96*math.sqrt((std**2)/2)))
```

# Run simulations

### Without batches

```
epsilon = 0.01
theta_ipa = 3.1
theta_sf = 3.1
theta_mvd = 3.1

iterations = []
thetas_ipa = []
thetas_sf = []
thetas_mvd = []
for i in range(10000):
  theta_ipa -= epsilon*(1 - G_IPA(theta_ipa))
  theta_sf -= epsilon*(1 - G_SF(theta_sf))
  theta_mvd -= epsilon*G_MVD(theta_mvd)

  if theta_ipa < 0.1:
    theta_ipa = 0.1
  if theta_sf < 0.1:
    theta_sf = 0.1
  if theta_mvd < 0.1:
    theta_mvd = 0.1

  iterations.append(i)
  thetas_ipa.append(theta_ipa)
  thetas_sf.append(theta_sf)
  thetas_mvd.append(theta_mvd)

plt.figure(figsize=(16, 8))
plt.plot(iterations, thetas_ipa, iterations, thetas_sf, iterations, thetas_mvd)
plt.xlabel('iteration')
plt.ylabel('theta')
plt.legend(['IPA', 'SF', 'MVD'])
plt.show()
```
